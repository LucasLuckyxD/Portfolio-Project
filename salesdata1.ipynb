```python
 import pandas as pd 
 import numpy as np
 import seaborn as sns
```

# Task 1 Merging 12 Dataset into one single file

```python
file_name = ['January', 'february', 'march', 'April', 'may', 'june', 'july', 'August', 'september', 'oktober', 'november', 'December']
```

```python
all_months_data = pd.DataFrame()
for x in file_name:
  df1 = pd.read_csv('/content/drive/MyDrive/Dataset Pandas/Sales_Order_1/'+x+'.csv')
  all_months_data = pd.concat([all_months_data,df1])
```

```python
all_months_data
```

```python
all_months_data.columns = all_months_data.columns.str.replace(' ','_')
```

```python
all_months_data['Order_Date'].str[0:2].unique()
```

By using the Unique Function, we can conclude that the merging of the dataset has been completed. With the Unique function we can also conclude that there are a few NaN values that we need to clean.

# Task 2 Cleaning The Data

```python
 filter_order_date_1 = all_months_data['Order_Date'].str[0:2] == 'Or'
 all_months_data[filter_order_date_1]
```

```python
all_months_data.dropna(axis='index',how = 'any',subset=['Order_ID', 'Product','Purchase_Address'])
```

```python
df_after_drop_na = all_months_data.copy()
df_after_drop_na.replace('Order Date',np.nan, inplace=True)
```

```python
df_after_drop_na.dropna(axis='index',how = 'any',subset=['Order_ID', 'Product','Purchase_Address', 'Order_Date'], inplace= True)
```

```python
df_after_drop_na['Order_Date'].str[0:2].unique()
```

```python
df_after_drop_na.dtypes
```

```python
df_after_drop_na['Quantity_Ordered'] = df_after_drop_na['Quantity_Ordered'].astype(float)
df_after_drop_na['Price_Each'] = df_after_drop_na['Price_Each'].astype(float)
```

```python
df_after_drop_na.dtypes
```

# Task 3 Best Month Sales and Visualisation

```python
df_copy_1 = df_after_drop_na.copy()
```

```python
df_copy_1['Income'] = df_copy_1['Quantity_Ordered']*df_copy_1['Price_Each']
df_copy_1
```

```python
df_copy_1['Month'] = df_copy_1['Order_Date'].str[0:2]
```

```python
df_copy_1['Month'] = df_copy_1['Month'].astype(int)
df_copy_1.dtypes
```

```python
df_copy_1['Month'] = df_copy_1['Month'].replace({12:'December',11:'November',10:'October',9:'September',8:'Agustus',7:'July',6:'June',5:'May',4:'April',3:'March',2:'February',1:'January'})
df_copy_1
```

```python
month_group = df_copy_1.groupby(['Month'])
month_group.get_group('March')
```

```python
month_group_income = month_group['Income'].sum()
```

```python
month_group_income
```

```python
new_df = pd.DataFrame({'Income':month_group_income}).reset_index()
new_df
```

```python
sns.barplot(y='Income',x='Month',data=new_df)
```

```python
sns.catplot(x='Income',y='Month',data=new_df, kind='bar')
```
